```python
import pandas as pd
import seaborn as sns
```

```python
gold = pd.read_csv('E_20201109_xauusd_price.csv')
```

```python
gold['date'] = pd.to_datetime(gold['date'])
```

```python
gold['XAUM1 Close'] = gold['XAUUSD Close']/ gold['M1 USD (B)']
gold.columns
```

```python
gold.describe()
```

```python
gold[['XAUUSD Close','M1 USD (B)', 'M2 USD (B)', 'XAUM1 Close']].describe()
```

### Gold Spot price in USD [1980-2020]

```python
sns.lineplot(data=gold, x='date', y='XAUUSD Close')
```

### Gold Spot price in USD vs M1 [1980-2020]

```python
sns.lineplot(data=gold, x='date', y='XAUUSD Close')
sns.lineplot(data=gold, x='date', y='M1 USD (B)')
```

### Gold Spot price in USD vs M1 and M2 in USD (B)[1980-2020]

```python
ax = gold.plot(x='date', y = 'XAUUSD Close')
sns.lineplot(data=gold, x='date', y='XAUUSD Close')
sns.lineplot(data=gold, x='date', y='M1 USD (B)')
sns.lineplot(data=gold, x='date', y='M2 USD (B)')
```

### Gold Spot in USD / M1 in USD (B)

```python
sns.lineplot(data=gold, x='date', y='XAUM1 Close')
```

### Gold Spot rolling 3 months volatility

```python
sns.lineplot(data=gold, x='date', y= gold['XAUUSD Vol  (3M)'])
```

```python
sns.lineplot(data=gold, x='date', y= gold['XAUUSD Vol  (6M)'])
```
