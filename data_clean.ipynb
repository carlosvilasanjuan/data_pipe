```python
import pandas as pd
import re
```

```python
gold = pd.read_csv('koyfin_20201107_xauusd_price.csv')
```

## Data Cleaning 

```python
gold
```

```python
# removing end space
gold['XAUUSD Open'] = [re.sub(' ', '', ele) for ele in gold['XAUUSD Open']]
gold['XAUUSD Low'] = [re.sub(' ', '', ele) for ele in gold['XAUUSD Low']]
gold['XAUUSD High']= [re.sub(' ', '', ele) for ele in gold['XAUUSD High']]
gold['XAUUSD Close'] =[re.sub(' ', '', ele) for ele in gold['XAUUSD Close']]
gold['XAUUSD Adj. Close'] =[re.sub(' ', '', ele) for ele in gold['XAUUSD Adj. Close']]

# removing comas
gold['XAUUSD Open'] = [re.sub(',','', ele) for ele in gold['XAUUSD Open']]
gold['XAUUSD Low'] = [re.sub(',','', ele) for ele in gold['XAUUSD Low']]
gold['XAUUSD High']= [re.sub(',','', ele) for ele in gold['XAUUSD High']]
gold['XAUUSD Close'] =[re.sub(',','', ele) for ele in gold['XAUUSD Close']]
gold['XAUUSD Adj. Close'] =[re.sub(',','', ele) for ele in gold['XAUUSD Adj. Close']]
```

```python
# Changing dtypes for price and dates and replacing ' Date' by 'Date' as name of the column
gold[['XAUUSD Open', 'XAUUSD Low', 'XAUUSD High', 'XAUUSD Close', 'XAUUSD Adj. Close']] = gold[['XAUUSD Open', 'XAUUSD Low', 'XAUUSD High', 'XAUUSD Close', 'XAUUSD Adj. Close']].astype(float)
gold
```

```python
gold['date'] = pd.to_datetime(gold[' Date'])
gold.index = gold.date 
gold = gold.drop(columns = ['date', ' Date', 'XAUUSD Volume'])
```

```python
gold.to_csv('C_20201107_xauusd_price.csv')
```
